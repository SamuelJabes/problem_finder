```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(
    model="text-embedding-3-large",
    dimensions=3072
    # With the `text-embedding-3` class
    # of models, you can specify the size
    # of the embeddings you want returned.
    # dimensions=1024
)
```

### Loading comments

```python
import json

with open("../data/comments/comments_with_problems.json", "r", encoding="utf-8") as file:
	comments_and_posts = json.load(file)

comments_and_posts
```

```python
reddit_embeddings = embeddings.embed_documents([c for c in comments_and_posts if c != None])
```

```python
with open("../data/embeddings/reddit_embeddings.json", "w", encoding="utf-8") as file:
    json.dump(reddit_embeddings, file)
```
