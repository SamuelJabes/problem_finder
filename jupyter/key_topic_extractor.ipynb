```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser

from dotenv import load_dotenv
import os

# Carregar variáveis de ambiente do arquivo .env
load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

# <Prompt>
key_topic_prompt = ChatPromptTemplate.from_messages(
        [
            (
                "system",
                """Você é um assistente pessoal que vai ler vários comentários e posts do Reddit que estão em um mesmo cluster semântico.
                   Os comentários/posts revelam algum teor problemático para o usuário que comenta/posta. Dentro desse cluster, a ideia é que as dores dos usuários sejam similares.
                   Assim, a sua tarefa é extrair as principais dores (os principais problemas) que os usuários estão enfrentando e que são comuns entre eles. Além disso, você deve
                   também extrair um tópico geral e abrangente que resuma o cluster, que diga qual cluster é esse, pois eles foram obtidos a partir de um algoritmo de clusterização como K-Means, logo,
                   precisamos dar um significado para esse cluster.
                   
                   Use o padrão abaixo para responder:
                   
                   [Nome do cluster]
                   [Início da lista de tópicos]
                    1. Topico 1
                        - Descrição do tópico 1
                    2. Topico 2
                        - Descrição do tópico 2
                    ...
                    [Fim da lista de tópicos]
                    
                    OBS1: Vale salientar que alguns comentários/posts não vão ter sentido, ou seja, não vão revelar nenhum problema. Ainda sim, extraia as dores comuns.
                    OBS2: Seria importante ter até 5 tópicos.
        """,
            ),
            (
                "human",
                """
                Aqui estão alguns comentários/posts do Reddit que revelam algum ou alguns problemas que os usuários estão enfrentando ou já enfrentaram:
                
                Comentários/posts: {comentarios_posts}
                
                (Cluster referente: {cluster_number})
                
                Por favor, me ajude a identificar os problemas comuns (os principais problemas) entre esses comentários/posts. Estou querendo encontrar alguma oportunidade para empreender e criar um produto
                ou serviço que resolva algum desses problemas.
        """,
            ),
        ]
    )

# <LLM>
llm = ChatOpenAI(
    model="gpt-4o-2024-08-06"
)

# <Chain>
chain = key_topic_prompt | llm | StrOutputParser()
    
```

```python
import json
input_path = '../data/clusters/reddit_clusters.json'

with open(input_path, 'r', encoding='utf-8') as f:
    clusters = json.load(f)
```

```python
for number_cluster in clusters:
    comentarios_posts_from_cluster = clusters[number_cluster]
    
    # Executar a cadeia de prompts
    result = chain.invoke({
        "comentarios_posts": comentarios_posts_from_cluster,
        "cluster_number": number_cluster
    })
    
    # Exibir o resultado
    print(f"{result}\n\n")
```

[Desafios do Trabalho e Rotina em Ambientes Híbridos e Remotos]
1. Problemas de Produtividade e Concentração
   - Muitos usuários relatam dificuldades em manter a produtividade e concentração ao trabalhar de casa (WFH - Work From Home). Questões como procrastinação, dificuldade de separar trabalho e vida pessoal, e a falta de disciplina sem a supervisão direta são comuns.

   - O ambiente de trabalho, seja em casa ou no escritório, tem grande impacto na produtividade. No escritório, distrações como fofocas e interações sociais desnecessárias são mencionadas. Já no home office, a falta de separação entre trabalho e descanso é um problema frequente.

3. Desafios da Estrutura Organizacional
   - Comenta-se sobre a ineficácia e o descontentamento com a estrutura atual de empresas, especialmente em relação ao suporte do RH e a gestão de equipes remotas. Além disso, há críticas aos modelos tradicionais e inflexíveis de horários, especialmente para pessoas com ADHD ou que são notívagas.

4. Opiniões Polarizadas sobre WFH vs. Presencial
   - Existe uma divisão clara nas preferências entre trabalho remoto e presencial. Algumas pessoas preferem WFH por menos distrações e flexibilidade, enquanto outras preferem o escritório para foco e socialização. Isso gera tensões em ambientes de trabalho híbridos.

5. Gestão de Tempo e Técnicas de Produtividade
   - Há frustrações com técnicas de produtividade como Pomodoro e desejos por hacks que funcionem para pessoas com estilos de vida diferentes, como as que são mais ativas à noite. Além disso, existem relatos de que ferramentas ou métodos tradicionais não atendem bem a pessoas com ADHD.


[Cluster sobre Problemas com Cobranças Médicas Inesperadas]

1. Bait-and-Switch em Cobranças Médicas
   - Muitos usuários relataram experiências em que serviços médicos foram inicialmente anunciados como gratuitos (como triagens ou consultas de rotina), mas acabaram resultando em cobranças significativas sem aviso prévio. Isso é frequentemente descrito como uma prática de "bait-and-switch", onde o serviço gratuito se transforma em um serviço pago quando os pacientes discutem outros problemas de saúde ou são encaminhados para tratamentos adicionais.

2. Falta de Transparência nas Cobranças
   - Os pacientes sentem-se enganados pela falta de clareza sobre os custos dos serviços médicos antes da consulta ou tratamento. Muitas vezes, eles não recebem informações claras sobre como as cobranças são feitas, resultando em surpresas financeiras.

3. Dificuldade em Resolver Disputas de Cobrança
   - Após receberem cobranças inesperadas, muitos pacientes enfrentam dificuldades para resolver disputas com clínicas, consultórios e seguradoras. O processo para corrigir códigos de cobrança errados ou contestar cobranças injustas pode ser complicado e estressante.

4. Impacto Negativo no Acesso a Cuidados de Saúde
   - Essas práticas e cobranças inesperadas desencorajam pacientes de buscar cuidados médicos futuros por medo de custos ocultos, afetando negativamente o acesso e a continuidade dos cuidados de saúde.

5. Necessidade de Educação sobre Seguros de Saúde
   - Há uma falta de entendimento entre os pacientes sobre como funcionam as coberturas de seguros e as práticas de cobrança no sistema de saúde, indicando uma necessidade de educação e recursos para ajudar os pacientes a navegar eficazmente no sistema de saúde.

[Cluster: Desafios Operacionais em Hotéis]

1. Uso de Tecnologia Antiquada
   - Muitos hotéis estão operando com computadores antigos e software desatualizado ou mal projetado, o que torna os sistemas lentos e ineficientes, causando frustração tanto para os funcionários quanto para os hóspedes.

2. Complexidade e Inflexibilidade nos Processos de Check-in
   - O processo de check-in é frequentemente demorado devido a sistemas complicados que requerem múltiplas verificações e entradas manuais de dados, exacerbadas pelo uso de tecnologia antiga.

3. Integração Falha com Sistemas de Reservas de Terceiros
   - Reservas feitas por sistemas de terceiros nem sempre transmitem todas as informações necessárias, como dados de contato ou métodos de pagamento, exigindo ajustes manuais pelos funcionários do hotel.

4. Necessidade de Multitarefa pela Equipe de Atendimento
   - Os recepcionistas estão sobrecarregados com várias tarefas simultâneas, como gerir reservas, responder a emails e coordenar com outras equipes, o que pode impactar a eficiência do atendimento.

5. Gestão de Preferências e Pedidos dos Hóspedes
   - A personalização do atendimento ao hóspede, como solicitações específicas de quarto ou resolução de conflitos de reserva, adiciona uma camada extra de complexidade ao processo, exigindo sistemas mais ágeis e uma melhor comunicação interna.

Esses problemas sugerem um mercado potencial para soluções tecnológicas mais modernas e integradas que possam otimizar os processos de check-in e a gestão de reservas em hotéis.
